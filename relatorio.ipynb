```python
def carregar_todas_acoes_BR():
  df1 = pd.read_excel(FONTE_DE_DADOS_1)

  df1.columns.values[0] = "Papel"
  df1 = df1.rename(columns=lambda x: x.replace(' ',''))

  HEADERS = {
          'User-Agent': 'Mozilla/5.0 (iPad; CPU OS 12_2 like Mac OS X) Applewebkit/605.1.15 (KHTML, Like Gecko) Mobile/15E148'
           }
  r = requests.get(FONTE_DE_DADOS_2, headers=HEADERS)

  output = str(r.content).replace("%", "")

  tables = pd.read_html(output, header=0, decimal=',', thousands='.')
  df2 = tables[0]

  df2.columns = [ 'Papel', 'Cotacao', 'P/L', 'P/VP', 'PSR', 'DY', 'P/ATIVOS', 'P/CAP. GIRO', 'P/EBIT',
                'P. AT CIR. LIQ', 'EV/EBIT', 'EV/EBITDA', 'MARGEM EBIT', 'MARG. LIQUIDA', 'LIQ. CORRENTE','ROIC',
                'ROE',
                'Liq.2meses', 'Patrim.Liq', 'Div.Brut/Patrim', 'Cresc.Rec.5a' ]

  merged_df = pd.merge(df1, df2, on='Papel', suffixes=('_left', '_right'), how='inner')

  merged_df = merged_df.drop(columns=merged_df.filter(like= '_left').columns)

  colunas = merged_df.columns

  novas_colunas = [coluna.replace("_rigth", "") for coluna in colunas]
  merged_df.columns = novas_colunas

  merged_df.fillna(0.0, inplace=True)

  df3 = pd.read_excel(FONTE_DE_DADOS_3)

  merged_df = pd.merge(merged_df, df3, on='Papel')

  return merged_df

```

```python
def get_viGraham(stock):
    try:
        return math.sqrt(22.5 * stock['LPA'] * stock['VPA'])
    except:
        return 0.0
```

```python
def get_viBarsi(stock):
    return (((stock['DY'] * stock['Cotacao']) / 100)  / 0.06)
```

```python
def fundamentalista(stock):
    pontos=0

    try:
         if (stock['Cotacao']<=get_viGraham(stock)):
             pontos+=1

          if (stock['P/L']<=20 and stock['P/L']>0):
             pontos+=1

          if (stock['ROE']>=10):
              pontos+=1
          if (stock['Cotacao']<=get_viBarsi(stock)):
              pontos+=1

          if ((stock['Cotacao']/stock['VPA'])<2)
              pontos+=1

          if (stock['MARGEMBRUTA'] >=40):
              pontos+=1

          if (((stock['Cresc.Rec.5a'] * stock['DY']) / stock['P/L']) >=1):
               pontos+=1

          return pontos
      except:
          return pontos
```
